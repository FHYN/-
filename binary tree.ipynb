```python
#Define binary treenode
class TreeNode:
    def __init__(self, x=-1): #默认为-1，用于初始化建树过程
        self.val = x
        self.left = None
        self.right = None

#build a tree, tranverse tree
class Tree():
    def __init__(self):
        self.root = None #根节点，初始化赋值-1
        self.queue = [] #层次遍历建树使用的队列
        self.ans = [] #记录遍历输出序列
        self.flag = 0 #0代表创建左孩子，1代表创建右孩子
    
    def clear_ans(self):
        self.ans = []
    
    def add(self, val): #层次遍历方式建立“完全”二叉树,None代表空节点
        if val is None:
            node = None
        else:
            node = TreeNode(val)
        #判断根节点是否为空
        if self.root is None:
            self.root = node
            self.queue.append(self.root)
        else:
            treenode = self.queue[0] #获取需要创建左右孩子的节点，类似于层次遍历二叉树的方式建树，通过flag标志位进行判断
            if self.flag == 0 and node is None:
                self.flag = 1
            elif self.flag == 1 and node is None:
                self.flag = 0
                self.queue.pop(0)
            elif self.flag == 0 and node is not None:
                treenode.left = node
                self.queue.append(treenode.left)
                self.flag = 1
            elif self.flag == 1 and node is not None:
                treenode.right = node
                self.queue.append(treenode.right)
                self.flag = 0
                self.queue.pop(0)
                
    def pre_recursive(self, root):
        if root is None:
            return None
        else:
            self.ans.append(root.val)
            self.pre_recursive(root.left)
            self.pre_recursive(root.right)
            
    def in_recursive(self, root):
        if root is None:
            return None
        else:
            self.in_recursive(root.left)
            self.ans.append(root.val)
            self.in_recursive(root.right)
            
    def post_recursive(self, root):
        if root is None:
            return None
        else:
            self.post_recursive(root.left)
            self.post_recursive(root.right)
            self.ans.append(root.val)
            
    def pre_traverse(self, root):
        if root is None:
            return None
        stack = []
        node = root
        while node or stack: #节点存在且堆栈非空
            while node:      #一直找左子树
                self.ans.append(node.val)
                stack.append(node)
                node = node.left
            node = stack.pop(-1)
            node = node.right
            
    def in_traverse(self, root):
        if root is None:
            return None
        stack = []
        node = root
        while node or stack: #节点存在且堆栈为空
            while node:      #一直找左子树
                stack.append(node)
                node = node.left
            node = stack.pop(-1)
            self.ans.append(node.val)
            node = node.right
            
    def post_traverse(self, root): #后序遍历相当于反向先序遍历（先右后左），再逆序输出
        if root is None:
            return None
        stack = []
        node = root
        while node or stack: #反向先序遍历
            while node:
                self.ans.append(node.val)
                stack.append(node)
                node = node.right
            node = stack.pop(-1)
            node = node.left
        self.ans = list(reversed(self.ans))    
        
    def level_traverse(self, root):
        if root is None:
            return None
        queue = []
        node = root
        while node or queue:
            if node:
                self.ans.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
                node = queue.pop(0) if queue else None
```

```python
lis = [1,2,3,None,1,3,2]
tree = Tree()
for i in lis:
    tree.add(i)
    
tree.pre_recursive(tree.root)
print("递归先序遍历结果:",tree.ans)
tree.clear_ans()
tree.in_recursive(tree.root)
print("递归中序遍历结果:",tree.ans)
tree.clear_ans()
tree.post_recursive(tree.root)
print("递归后续遍历结果:",tree.ans)
tree.clear_ans()

tree.pre_traverse(tree.root)
print("非递归先序遍历结果:",tree.ans)
tree.clear_ans()
tree.in_traverse(tree.root)
print("非递归中序遍历结果:",tree.ans)
tree.clear_ans()
tree.post_traverse(tree.root)
print("非递归后序遍历结果:",tree.ans)
tree.clear_ans()
tree.level_traverse(tree.root)
print("层次遍历结果:",tree.ans)
tree.clear_ans()
```
